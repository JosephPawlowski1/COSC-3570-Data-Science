```python
#imports that were used throughout the program
from datascience import *
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
```

```python
#Link to data set used: https://www.kaggle.com/daverosenman/ufc-ppv-sales
#loading data set into program
PPV_Sales_Table = Table.read_table('ufc_ppv_buys.csv')
```

```python
#Diplaying the data set in a readable table
PPV_Sales_Table
```

```python
#splitting the data set by year 
PPV_Sales_2001 = PPV_Sales_Table.where('Year',2001)
PPV_Sales_2002 = PPV_Sales_Table.where('Year',2002)
PPV_Sales_2003 = PPV_Sales_Table.where('Year',2003)
PPV_Sales_2004 = PPV_Sales_Table.where('Year',2004)
PPV_Sales_2005 = PPV_Sales_Table.where('Year',2005)
PPV_Sales_2006 = PPV_Sales_Table.where('Year',2006)
PPV_Sales_2007 = PPV_Sales_Table.where('Year',2007)
PPV_Sales_2008 = PPV_Sales_Table.where('Year',2008)
PPV_Sales_2009 = PPV_Sales_Table.where('Year',2009)
PPV_Sales_2010 = PPV_Sales_Table.where('Year',2010)
PPV_Sales_2011 = PPV_Sales_Table.where('Year',2011)
PPV_Sales_2012 = PPV_Sales_Table.where('Year',2012)
PPV_Sales_2013 = PPV_Sales_Table.where('Year',2013)
PPV_Sales_2014 = PPV_Sales_Table.where('Year',2014)
PPV_Sales_2015 = PPV_Sales_Table.where('Year',2015)
PPV_Sales_2016 = PPV_Sales_Table.where('Year',2016)
PPV_Sales_2017 = PPV_Sales_Table.where('Year',2017)
```

```python
#finding the average of each each year 
PPV_Sales_2001_Avg = np.mean(PPV_Sales_2001[6])
PPV_Sales_2002_Avg = np.mean(PPV_Sales_2002[6])
PPV_Sales_2003_Avg = np.mean(PPV_Sales_2003[6])
PPV_Sales_2004_Avg = np.mean(PPV_Sales_2004[6])
PPV_Sales_2005_Avg = np.mean(PPV_Sales_2005[6])
PPV_Sales_2006_Avg = np.mean(PPV_Sales_2006[6])
PPV_Sales_2007_Avg = np.mean(PPV_Sales_2007[6])
PPV_Sales_2008_Avg = np.mean(PPV_Sales_2008[6])
PPV_Sales_2009_Avg = np.mean(PPV_Sales_2009[6])
PPV_Sales_2010_Avg = np.mean(PPV_Sales_2010[6])
PPV_Sales_2011_Avg = np.mean(PPV_Sales_2011[6])
PPV_Sales_2012_Avg = np.mean(PPV_Sales_2012[6])
PPV_Sales_2013_Avg = np.mean(PPV_Sales_2013[6])
PPV_Sales_2014_Avg = np.mean(PPV_Sales_2014[6])
PPV_Sales_2015_Avg = np.mean(PPV_Sales_2015[6])
PPV_Sales_2016_Avg = np.mean(PPV_Sales_2016[6])
PPV_Sales_2017_Avg = np.mean(PPV_Sales_2017[6]) 
```

```python
#creating two arrays: one to display the year in a easer way and of all of the
#average PPV sales.
years = ['01','02','03','04','05','06','07','08'
         ,'09','10','11','12','13','14','15','16'
        ,'17']
PPV_sales_avg = [PPV_Sales_2001_Avg,PPV_Sales_2002_Avg,PPV_Sales_2003_Avg,PPV_Sales_2004_Avg,
                PPV_Sales_2005_Avg,PPV_Sales_2006_Avg,PPV_Sales_2007_Avg,PPV_Sales_2008_Avg,
                PPV_Sales_2009_Avg,PPV_Sales_2010_Avg,PPV_Sales_2011_Avg,PPV_Sales_2012_Avg,
                PPV_Sales_2013_Avg,PPV_Sales_2014_Avg,PPV_Sales_2015_Avg,PPV_Sales_2016_Avg,
                PPV_Sales_2017_Avg]
```

```python
#creating a bar graph that displays the average PPV sales of each year from 2009 to 2017
plots.figure(figsize=(8,8))
PPV_chart = plots.subplot()
PPV_chart.bar(years,PPV_sales_avg,color = 'green')
PPV_chart.set_xlabel('Months')
PPV_chart.set_ylabel('PPV Sales')
PPV_chart.set_title('UFC PPV Sales 2001-2017')
```
